```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

### Store CSV into DataFrame

```python
# Table 1 - CSV data
csv_file = "Resources/housing_data_2018.csv"
housing_data_2018_df = pd.read_csv(csv_file)
housing_data_2018_df.head()
```

```python
# replace the columns
housing_data_2018_df.rename(columns={"1/31/2018":"Jan_2018",
                                     "2/28/2018":"Feb_2018",
                                     "3/31/2018":"Mar_2018",
                                     "4/30/2018":"Apr_2018",
                                     "5/31/2018":"May_2018",
                                     "6/30/2018":"Jun_2018",
                                     "7/31/2018":"Jul_2018",
                                     "8/31/2018":"Aug_2018",
                                     "9/30/2018":"Sep_2018",
                                     "10/31/2018":"Oct_2018",
                                     "11/30/2018":"Nov_2018",
                                     "12/31/2018":"Dec_2018",}, inplace=True)

housing_data_2018_df
```

```python
housing_data_2018_df['Jan_2018'] = housing_data_2018_df['Jan_2018'].astype('Int64')
housing_data_2018_df['Feb_2018'] = housing_data_2018_df['Feb_2018'].astype('Int64')
housing_data_2018_df
```

```python
housing_data_2018_df.dtypes
```

### Create new data with select columns

```python
new_housing_data_2018_df = housing_data_2018_df[['StateName', 
                                                 'RegionID', 
                                                 'Jan_2018',
                                                 'Feb_2018',
                                                 'Mar_2018',
                                                 'Apr_2018',
                                                 'May_2018',
                                                 'Jun_2018',
                                                 'Jul_2018',
                                                 'Aug_2018',
                                                 'Sep_2018',
                                                 'Oct_2018',
                                                 'Nov_2018',
                                                 'Dec_2018',]]
new_housing_data_2018_df.head()
```

```python
# Table 2 - CSV Census Data
csv_file = "Resources/census_data_states_2018.csv"
census_data_2018_df = pd.read_csv(csv_file)
census_data_2018_df.head()
```

```python
census_data_2018_df.dtypes
```

```python
census_data_2018_df.rename(columns={"Name":"StateName"}, inplace=True)
census_data_2018_df.head()
```

```python
# Table 3 - CSV data "Mean Pending Data"
csv_file = "Resources/mean_pending_data_2018.csv"
pending_data_2018_df = pd.read_csv(csv_file)
pending_data_2018_df = pending_data_2018_df.dropna()
pending_data_2018_df.head()
```

```python
# rename columns
pending_data_2018_df.rename(columns={"Name":"StateName"}, inplace=True)
pending_data_2018_df.head(5)
```

```python
pending_data_2018_df.rename(columns={"1/31/2018":"Jan_2018",
                                     "2/28/2018":"Feb_2018",
                                     "3/31/2018":"Mar_2018",
                                     "4/30/2018":"Apr_2018",
                                     "5/31/2018":"May_2018",
                                     "6/30/2018":"Jun_2018",
                                     "7/31/2018":"Jul_2018",
                                     "8/31/2018":"Aug_2018",
                                     "9/30/2018":"Sep_2018",
                                     "10/31/2018":"Oct_2018",
                                     "11/30/2018":"Nov_2018",
                                     "12/31/2018":"Dec_2018",}, inplace=True)

pending_data_2018_df.head(5)
```

```python
pending_data_2018_df['Jan_2018'] = pending_data_2018_df['Jan_2018'].astype('Int64')
pending_data_2018_df['Feb_2018'] = pending_data_2018_df['Feb_2018'].astype('Int64')
pending_data_2018_df
```

```python
new_pending_data_2018_df = pending_data_2018_df[['StateName', 
                                                 'RegionID', 
                                                 'Jan_2018',
                                                 'Feb_2018',
                                                 'Mar_2018',
                                                 'Apr_2018',
                                                 'May_2018',
                                                 'Jun_2018',
                                                 'Jul_2018',
                                                 'Aug_2018',
                                                 'Sep_2018',
                                                 'Oct_2018',
                                                 'Nov_2018',
                                                 'Dec_2018',]]
new_pending_data_2018_df.head()
```

```python
new_pending_data_2018_df.dtypes
```
